Напишите код на Python, реализующий расчет длины вектора, заданного его координатами. (в программе)

```python
import numpy as np

def mod_vector(v):
    if type(v) != 'numpy.ndarray':
        v = np.array(v)
        
    v = v**2
    return np.sqrt(v.sum())

a = [3,5]
print(mod_vector(a))
```

## 1. Задание (в программе)
Нарисуйте график функции:
y(x) = k∙cos(x – a) + b
для некоторых (2-3 различных) значений параметров k, a, b


```python
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
def f(x,a,b,k):    
    return k*np.cos(x-a)+b
    
x = np.linspace(-5,5,101)
y1 = f(x,1,0,1)
y2 = f(x,3,2,8)
y3 = f(x,4,6,0)

plt.plot(x,y1,label='Y1')
plt.plot(x,y2,label='Y2')
plt.plot(x,y3,label='Y3')

plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

### 1. Напишите код, который будет переводить полярные координаты в декартовы.

```python
def dec2radial(x,y):
    r = np.sqrt(x*x+y*y)
    a = np.arctan(y/x)
    return (r,a)

print(dec2radial(6,8))
```

# Задание 4

Решите систему уравнений:

y = x2 – 1

exp(x) + x∙(1 – y) = 1

```python
from scipy.optimize import fsolve
import math

def my_equations(p):
    x, y = p
    
    return (x**2-1-y, math.exp(x) + x*(1-y)-1)

x, y =  fsolve(my_equations, (10, 10))
print (x, y)

x, y =  fsolve(my_equations, (1, 1))
print (x, y)
```
