# 4. Задание (в программе):
Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.


```python
import numpy as np
import pandas as pd
import math 
from matplotlib import pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
k1 = 1
k2 = 4
```

```python
x = np.linspace(-10,10,300)
```

```python
plt.plot(x,np.cos(k1*x),label=f'K={k1}')
plt.plot(x,np.cos(k2*x),label=f"K={k2}")
plt.title('y(k,x)=cos(k∙x) с разным значением k')
plt.legend()
plt.show()
```
